# Project Data Visualisation

```python
import pandas as pd
import folium
import pyspark


# Data inladen
ziekenhuizen = pd.read_csv('data/Ziekenhuis_overzicht.csv')
leeftijd = pd.read_csv('data/Leeftijd_per_wijk.csv')
wijkenranking = pd.read_csv('data/Wijkenranking.csv')

min_lon, max_lon = 4.35, 4.45
min_lat, max_lat = 51.18, 51.25

m = folium.Map(
    max_bounds=True,
    location=(51.23016715, 4.4161294643975015),
    zoom_start=13,
    min_zoom=13,
    max_lat=max_lat,
    min_lat=min_lat,
    min_lon=min_lon,
    max_lon=max_lon,
)

m
m.save('output/map.html')
```
